# **Mô hình dữ liệu quan hệ**
1. Thuộc tính (Attribute)
- Mô tả đặc trưng, tính chất riêng biệt của đối tượng cần được lưu trữ trong CSDL.
-**Tên gọi**: dãy ký tự
-**kiểu dữ liệu**: Số, chuỗi, ...
-**Miền giá trị**: Dom(TenThuocTinh).
Vd: Dom(GioiTinh) = ('Nam','Nu').

2.Quan hệ (Relation).
- Quan hệ là một tập hữu hạn các thuộc tính.
- Quan hệ $R$ có $n$ ngôi được định nghĩa trên tập thuộc tính $R(A_1, A_2,...,A_n)$ và kèm theo một tân từ để xác định mối quan hệ giữa các thuộc tính $A_j$.
- Ký hiệu $R(A_1,A_2,...,A_n)$
- $R$: Tên quan hệ
- $R^+ = (A_1, A_2,...,A_n)$: Tập thuộc tính của $R$.

Vd: **SinhVien**(MaSV,HoTen,GioiTinh,MaLop).

3. Bộ giá trị (Tuple).

- Là các thông tin của một đối tượng thuộc quan hệ.
- Còn được gọi mẫu tin (record), dòng (row).
-Một bộ của quan hệ $R(A_1,A_2,...,A_n)$ là $r=(a_1,a_2,...,a_n)$ với $∀a_j∈DOM(A_j)$

4. Thể hiện của quan hệ

- Là tập hợp các bộ giá trị của quan hệ tại một thời điểm
- Ký hiệu: $T_R$

5. Tân từ
- Là một quy tắc dùng để mô tả một quan hệ.
- Làm rõ ngữ nghĩa, sự liên hệ giữa các thuộc tính trong quan hệ.

- Ký hiệu: $||R||$


6. Phép chiếu
- **Chiếu của một quan hệ lên tập thuộc tính**:
- Dùng để trích giá trị của một số thuộc tính trogn danh sách các thuộc tính của quan hệ.
- Ký hiệu: $R[X]$ hoặc $R.X$ - Phép chiếu của quan hệ $R$ lên tập thuộc tính $X$.

- **Chiếu của một bộ lên tập thuộc tính**:
- Trích chọn các giá trị cụ thể của một bộ theo các thuộc tính đưuọc chỉ ra trong danh sách thuộc tính của một quan hệ.
- Ký hiệu: $t_R[X]$ - Chiếu của một bộ giá trị $t$ lên tập thuộc tính $X$ của quan hệ $R$.

7. Khóa

7.1 Siêu khóa (Super Key).
- Siêu khóa $SK$ là một tập con các thuộc tính cảu $R^+$ mà giá trị của chúng có thể phân biệt 2 bộ khác nhau trong một thể hiện $T_R$ bất kỳ.
- $∀t_1,t_2 ∈ T_R, t1[SK] ≠ t_2[SK] ⟺ SK$ là siêu khóa của R.

- Một quan hệ luôn có ít nhất một siêu khóa $(R^+)$ và có thể có nhiều siêu khóa.

7.2 Khóa (key)

- $K$ là khóa của quan hệ $R$ khi thỏa mãn 2 điều kiện:
- $K$ là một siêu khóa
-$K$ là siêu khóa nhỏ nhất (chứa ít thuộc tính nhất và khác rỗng):
- $¬∃ K_1 \subset K,K_1 ≠ ∅:K_1 $ là siêu khóa.
- Thuộc tính tham gia và một khóa gọi là thuộc tính khóa.

7.3 Khóa chính (Primary Key).
- Khi thiết kế CSDL, nếu quan hệ có nhiều hơn một khóa, ta chỉ được chọn một và gọi là khóa chính.

7.4 Khóa tương đương
- Các khóa còn lại gọi là khóa tương đương (không chọn làm khóa chính).

7.5 Khóa ngoại (Foreign Key).
- Cho 2 quan hệ $R(U),S(V)$ với $K_1 ⊆ U$ và là khóa chính của $R,K_2 ⊆V$.
- khi đó $K_2$ là khóa ngoại của $S$ tham chiếu đến khóa chính $K_1$ của $R$ nếu thỏa mãn:
- $K_1$ và $K_2$ có cùng số lượng thuộc tính và ngữ nghĩa của các thuộc tính trong $K_1$ và $K_2$ giống nhau.
- Giữa $R$ và $S$ tồn tại một mối quan hệ 1-n trên $K_1$ và $K_2$.
- $∀s∈S,!∃r∈R: r.K_1 = s.K_2$

8. Lược đồ quan hệ
- Lược đồ quan hệ mô tả cấu trúc của một quan hệ và các mối liên hệ giữa các thuộc tính trong quan hệ đó.
- Cấu trúc của một quan hệ là tập thuộc tính của quan hệ đó.

# **Phụ thuộc hàm**

Khái niệm:
- Phụ thuộc hàm trên quan hệ $R$ biểu diễn mối liên hệ giữa các tập thuộc tính trong $R$.
- Ký hiệu: $X \to Y$
- Định nghĩa:
  - Quan hệ $R(A_1,A_2,...,A_n), R^+ = {A_1,A_2,...,A_n}$
  - $X,Y ⊂ R^+$
  - $t1,t2$: 2 bộ bất kỳ trên $R$.
  - Ta nói: $X \to Y ⟺ (t1.X = t2.X ⟹ t1.Y = t2.Y)$

**Hệ luật dẫn Amstrong**
Quan hệ $R(A,B,C)$ và $F$ là tập PTH
- $ X → Y$ được suy ra từ F nếu bất kỳ bộ của quan hệ R thỏa F thì
cũng thỏa $X → Y$
- Ký hiệu: $F ╞ X → Y$

- Tính chất:
  - Tính phản xạ: $MaSV,TenSV \to TenSV$
  - Tính tăng trưởng: $MaSV \to TenSV$, $⟹MaSV,NgaySinh \to TenSV,NgaySinh$
  - Tính bắc cầu: $MaSV \to MaLop$, $MaLop \to TenLop$, $⟹ MaSV \to TenLop$
  - Tính kết hợp: $MaSV \to TenSV$, $MaSV \to GioiTinh$, $⟹MaSV \to TenSV,GioiTinh$
  - Tính phân rã: $MaSV \to TenSV,GioiTinh$, $⟹MaSV \to TenSV, MaSV \to GioiTinh$
  - Tính tựa bắc cầu: $MaSV \to MaLop, MaLop, MaMon \to MaGV$, $⟹MaSV, MaMon \to MaGV$

**Bao đóng**:
- Bao đóng của tập phụ thuộc hàm $F$:
  - Ký hiệu :$F^+$
  - Là tập tất cả các phụ thuộc hàm được suy ra từ $F$.
- Bao đóng của tập thuộc tính $X$ đối với tập phụ thuộc hàm $F$:
  - Ký hiệu: $X_F^+$
  - Là tập tất cả các thuộc tính A có thể suy dẫn từ X nhờ tập bao đóng của các phụ thuộc hàm $F^+$.
  - $X_F^+ = A ∈ R^+ | X \to A \in F^+ $

**Tìm bao đóng của tập thuộc tính X đối với tập phụ thuộc hàm F**:
- Input $(R,F), X ⊆ R^+$
- Output: $X_F^+$
- Bước 1: Tính dãy $X^{(0)},X^{(1)},...,X^{(i)}$
  - $X^{(0)} = X$
  - $X^{(i+1)} = X^{(i)} ∨ Z, ∃(Y \to Z) \in F(Y ⊆ X^{(i)})$, loại $(Y \to Z)$ ra khỏi $F$.
  - Dừng khi $X^{(i+1)} = X^{(i)}$ hoặc $X^{(i)}=R^+$
- Bước 2: Kết luận $X_F^+ = X^{(i)}$

Ví dụ: Cho $R(A,B,C,D,E,G,H)$ và tập PTH:
$F=$ {$f1: B \to A, f2: DA \to CE, f3: D \to H, f4: GH \to C, f5: AC \to D$}.
Tìm $AC_F^+$ và $AC \to E \in F^+$ hay không?

- $X^{(0)} = AC$
- từ f1 đến f4 không thỏa, f5 thỏa nên $X^{(1)} = ACD$

- xét $X^{(1)} = ACD$
- f1 không thoả
- f2 thỏa nên $X^{(1)} = ACDE$
- f3 thỏa nên $X^{(1)} = ACDEH$
- f4 không thỏa
- f5 đã thỏa

- xét $X^{(2)} = ACDEH$
- f1 không thoản
- f2,f3,f5 đã thỏa
- f4 không thỏa

Vậy: $AC_F^+ =  ACDEH$

$X \to Y \in F^+ ⟺ Y ⊆ X^+$

Vì $AC_F^+ = ACDEH$ và $E \in AC_F^+$ nên $AC \to E \in F^+$

**Phủ tối thiểu**
- Hai tập phủ tối thiểu tương đương nếu $F^+ = G^+ ⟺ G ≡ F$

- **Phụ thuộc hàm có thuộc tính vế trái dư thừa**:
  - Cho F là tập PTH trên R, $X \to Y \in F$ là PTH có thuộc tính vế trái dư thừa nếu: $∃A \in F| F = (X \to Y) ∪ ((X=A) \to Y) $
  - Ngược lại $Z \to Y$ là PTH thuộc tính vế trái không dư thừa.

Ví dụ: $R(A,B,C,D)$ và $F = (A \to B, BC \to D, C \to D)$. Khi đó $BC \to D$ là PTH có thuộc tính vế trái dư thừa do chỉ cần $C \to D$.

- **Thuật toán tìm PTH đầy đủ**:
  - Với mỗi PTH $X \to Y, X = A_1 A_2 ... A_n$ và $n ≥ 2$, đặt $Z=X$
  - Với mỗi $A_i$, thực hiện:
    - $tmp = Z \ A_i$
    - If $tmp \to Y \in (F = {X \to Y}^+ ⟹ Z = tmp$

Ví dụ: $R(A,B,C,D)$ và $F = (A \to B, BC \to D, C \to D)$.
- Xét $BC \to D$
  - Có $C_{F-{BC \to D}}^+ = CD$
  - Vì $D ⊆ C_F^+$ nên B là thuộc tính dư thừa -> Loại bỏ B
  - Vậy $F = A \to B, C \to D$

- **Tập PTH có vế phải một thuộc tính**: Mỗi tập PTH F đều tương đương với một tập PTH G mà vế phải của các PTH thuộc G chỉ gồm 1 thuộc tính.

- **PTH không dư thừa**:
  - F là tập PTH không dư thừa nế không tồn tại $F' ⊂ F | F' ≡ F$

- **Thuật toán loại những PTH dư thừa**: Với mỗi PTH $X \to Y \in F$, nếu $X \to Y$ là thành viên của $F - {X \to Y}$ thì loại $X \to Y$ khỏi $F$.

- **Phủ tối thiểu của một tập phụ thuộc hàm** (F) nếu:
  - F có thuộc tính vế trái không dư thừa
  - F có vế phải là một thuộc tính
  - F không dư thừa

- **Thuật toán tìm PTT của một tập phụ thuộc hàm**
  - B1: Phân rã các PTH có vế phải nhiều thuộc tính thành các PTH có vế phải một thuộc tính.
  - B2: Loại bỏ các thuộc tính vế trái dư thừa của các PTH
  - B3: Loại bỏ các PTH dư thừa

Ví dụ: $R(A,B,C,D)$ và $F = (A \to B, BC \to D, C \to D)$. Tìm phủ tối thiểu:
- B1: $F = f1: AB \to C, f2: AB \to D, f3: B \to C, f4: C \to D$
- B2: f3,f4 không xét vì có một thuộc tính ở vế trái.
- Xét $f1: AB \to C$
  - Nếu bỏ $A$ thì $B_{F-f1}^+ = BCD$ chứa $C$ nên bỏ $A$.
  - Nếu bỏ $B$ thì $A_{F-f1}^+ = A$ không chứa C, không bỏ B
  - $⟹ f1: B \to C$

- Xét $f2: AB \to D$
  - Nếu bỏ $A$ thì $B_{F-f2}^+ = BCD$ chứa $D$ nên bỏ $A$
  - Nếu bỏ $B$ thì $A_{F-f2}^+ = A$ không chứa $D$, không bỏ $B$
  - $ ⇒ B \to D $
Vậy: Phủ tối thiểu là $F = B \to C, C \to D$

- **Khóa**:
  - B1: Xác định tập nguồn $N$ chứa thuộc tính chỉ xuất hiện ở vế trái các PTH:
  - $N_F^+$:
    - $N_F^+ = R^+$: Khóa là N
    - Ngược lại, tiếp tục đến bước 2
  - B2: Xác định tập nguồn trung gian $TG$, chứa các thuộc tính xuất hiện ở cả vế trái và vế phải của các PTH
  - Xác định các tập con $X_i$ có thể có của tập $TG$.
  - Bước 3: $∀X_i ⊆ TG$ nếu $(N ∪ X_i)_F^+ = R^+$: $S_i = N ∪ X_i$, loại bỏ các tập $X_k: X_i ⊂ X_j$
  - Bước 4: Kết luận tập các khóa $K = {S_j}$

VD: Cho $R (A,B,C,D,E,G,H)$, tập $PTH = B \to A, DA \to CE, D \to H, GH \to C, AC \to D$. Tìm khóa.

B1: Tập nguồn $N = B,G$
- Ta có: $BG_F^+ = BGA ≠ R^+$ : BG không là khóa
B2: Tập trung gian $TG = A,C,D,H$
- Các tập con của $TG = A,C,D,H,AC,AD,AH,CD,CH,DH,ACD,ACH,ADH,CDH,ACDH$
- B3:

|-|-|-|-|

# **Dạng Chuẩn**

- Chuẩn hóa các quan hệ (Giảm tối đa trùng lặp thông tin, kiểm tra RBTV dễ dàng).

**Dạng chuẩn 1**
  - tất cả thuộc tính mang giá trị nguyên tố (không thể phân nhỏ).

**Dạng chuẩn 2**:
  - Đạt DC1
  - Các thuộc tính không khóa đều **Phụ thuộc đầy đủ** và Khóa
  - Kiểm tra DC2:
    - Tìm tất cả các khóa của R
    - Với mỗi khóa K, tìm $S_i^+$ với $S_i$ là tất cả các tập con thực sự của K.
    - Nếu tồn tại $S_i^+$ chứa thuộc tính không khóa thì R không đạt DC2, ngược lại Q đạt DC 2.

VD: $R(A,B,C,D)$, PTH: $F = A \to B, B \to DC$
- Tìm khóa của R:
  - Tập nguồn $N = A$
  - $A_F^+ = ABCD$ => lược đồ có khóa là $A$.

VD: Cho tập $SinhVien (MaSV, MaMH, TenSV, DiaChi, DIem)$, PTH $F = MaSV, MaMH \to Diem, MaSV \to TenSV, DiaChi$.

- Tập nguồn $N = MaSV, MaMH$
- $N_F^+ = MaSV,MaMH,TenSV,DiaChi,Diem$ nên $MaSV, MaMH$ là khóa

- Ta thấy $MaSV ⊂ K, MaSV \to TenSV, DiaChi$. $TenSV,DiaChi$ là thuộc tính không khóa, nên không phụ thuộc đầy đủ vào khóa. => SINHVIEN không đạt DC2.

=> tách thành 2 lược đồ, đạt DC2:
- $DANGKY(MaSV, MaMH, Diem).$
- $SinhVien(MaSV, TenSV,DiaChi).$

**Dạng chuẩn 3**:
  - Đạt DC2
  - Các thuộc tính không khóa không phụ thuộc bắc cầu vào khóa.
  - Kiểm tra DC3:
    - Tìm tất cả các khóa của R
    - Phân rã vế phải của các PTH trong F thành các PTH có vế phải một thuộc tính
    - Nếu mọi PTH $X \to Y \in F$ và $Y ∉ X$: X là siêu khóa (vế trái chỉ chứa 1 khóa) hoặc Y là thuộc tính khóa (vế phải là tập con của khóa) thì R đạt DC3.

**Dạng chuẩn Boyce Codd (BC)**:
- Nếu tát cả các PTH $X \to Y in F$ với $Y ∉ X$ đều có $X$ là siêu khóa.

VD: Cho $R (A, B, C, D, E, I)$, PTH $F = ACD \to EBI, CE \to AD$

- Khóa của R: $ACD,CE$
- Phân rã: $F = ACD \to E, ACD \to B, ACD \to I, CE \to A, CE \to D$

- Các PTH trong F đều có vế trái là một siêu khóa.
